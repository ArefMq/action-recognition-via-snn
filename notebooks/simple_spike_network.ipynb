# Simple Spiking Dense Network on MNIST

A two-layer LIF network trained on spike-encoded MNIST. Shows how to build and visualise a spiking network using `SpikingDenseLayer` with surrogate-gradient backprop.

```python
from spikenet.dataset.spiking_mnist import spiking_mnist
from spikenet.image_to_spike_convertor import SpikePlotter
from spikenet.network import Network

splt = SpikePlotter()
```

## Network

Two `SpikingDenseLayer` layers: input → 400 → 10. The final layer uses `MemRecMax` time reduction to collapse the temporal dimension into a class score. A single forward pass is run before training to confirm the shapes and inspect neuron activity.

```python
from spikenet.layers.spiking_base import TimeReduction
from spikenet.layers.spiking_dense import SpikingDenseLayer

data_x, data_y = spiking_mnist.sample()
net = (
    Network()
    .add_layer(SpikingDenseLayer, 400, in_features=data_x.shape[-1])
    .add_layer(SpikingDenseLayer, 10, time_reduction=TimeReduction.MemRecMax)
).build()


data_x = data_x.reshape(1, *data_x.shape)
net.forward(data_x)
net.plot_activity()
```

## Training

`.fit()` compiles the network, runs the epoch loop with surrogate-gradient backprop, and evaluates on the test split.

```python
net.fit(spiking_mnist)
```

## Post-training activity

Compare spiking activity before and after training — trained neurons should show sparser, more selective firing patterns.

```python
net.forward(data_x)
net.plot_activity()
```
