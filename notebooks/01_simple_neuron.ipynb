```python
from spikenet.layers import SpikingDenseLayer
```

```python
neuron = SpikingDenseLayer(out_features=1, in_features=10)
neuron.initialize_parameters()
neuron
```

```python
import torch

# Shape: (batch=1, time_steps=100, in_features=1)
# A constant input current — value controls how often the neuron fires
x = torch.bernoulli(torch.full((1, 100, 10), 0.1))
_ = neuron.forward(x)
```

```python
neuron.plot_activity()
neuron.plot_membrane()
```
